Data:

 郵便番号データダウンロード > 読み仮名データの促音・拗音を小書きで表記するもの(zip形式)

  ZIP code data download > Reading kana data with small letters for uromandatory and sulphuric sounds (zip format)

https://www.post.japanpost.jp/zipcode/dl/kogaki-zip.html

```python
import pandas as pd

# Read data frames
df = pd.read_csv("13TOKYO.CSV", header=None, encoding="shift_jis")
print(len(df))
print(df.columns.values)
```

```python
df.head()
```

```python
# Extract and display addresses where column “2” is "1600006
results = df[df[2] == 1600006]
print(results[[2,6,7,8]])
```

```python
# Extract and display addresses where column “8” is “Yotsuya”.
results = df[df[8] == "四谷"]
print(results[[2,6,7,8]])

# Extract and display addresses that contain the word “Yotsuya” in the “8” column.
results = df[df[8].str.contains("四谷")]
print(results[[2,6,7,8]])
```
