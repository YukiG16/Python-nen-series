```python
import pandas as pd

# Read data frames
df = pd.read_csv("200.csv",encoding="shift-jis")

print(len(df))
print(df.columns.values)
```

```python
# Fire Hydrant: List the location (latitude and longitude) of the fire hydrant.
hydrant = df[["緯度","経度"]].values
print(len(hydrant))
print(hydrant)
```

```python
import folium

# Make a map and write it down
m = folium.Map(location=[35.942957, 136.198863], zoom_start=16)
for data in hydrant:
    folium.Marker([data[0], data[1]]).add_to(m)
m
```

```python
# Read data frames
df = pd.read_csv("898.csv")

print(len(df))
print(df.columns.values)
```

```python
# List the locations (latitude and longitude) and stores where the stores are located
store = df[["緯度","経度","店舗名(日本語)"]].values
print(len(store))
print(store)
```

```python
# Make a map and write it down
m = folium.Map(location=[35.942957, 136.198863], zoom_start=16)
for data in store:
    folium.Marker([data[0], data[1]], tooltip=data[2]).add_to(m)
m
```
