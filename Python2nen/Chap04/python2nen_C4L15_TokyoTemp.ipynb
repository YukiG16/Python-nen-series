```python
!pip install japanize_matplotlib
import pandas as pd
import matplotlib.pyplot as plt
import japanize_matplotlib
```

```python
# Read data frames
df1 = pd.read_csv("Preview_20250407024350MeanT.csv", skiprows=1, index_col="時点")
df2 = pd.read_csv("Preview_20250407024333MaxT.csv", skiprows=1, index_col="時点")
df3 = pd.read_csv("Preview_20250407024315MinT.csv", skiprows=1, index_col="時点")

print(df1.columns.values)
print(df2.columns.values)
print(df3.columns.values)
```

```python
# Line graph with average temperature
df1["東京都"].plot(title='年平均気温【℃】')
plt.show()
```

```python
# Line graph with average temperature
df1["東京都"].plot(title='年平均気温【℃】')
plt.ylim(-10,40)
plt.show()
```

```python
# Overlay three graphs on top of each other
df1["東京都"].plot(label="年平均気温【℃】")
df2["東京都"].plot(label="最高気温（日最高気温の月平均の最高値）【℃】")
df3["東京都"].plot(label="最低気温（日最低気温の月平均の最低値）【℃】")
plt.ylim(-10,40)
plt.legend(loc="lower right")
plt.show()
```
