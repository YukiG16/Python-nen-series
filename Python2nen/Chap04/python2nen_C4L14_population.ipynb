https://www.e-stat.go.jp/stat-search/database?page=1&toukei=00200524&tstat=000000090001

```python
import pandas as pd
# Read data frames
df = pd.read_csv("/content/FEH_00200524_210901131442.csv", index_col="全国・都道府県", encoding="shift_jis")
print(len(df))
print(df.columns.values)
```

```python
!pip install japanize-matplotlib
import matplotlib.pyplot as plt
import japanize_matplotlib

# Create and display a bar chart sorted by population with column data for the year 2008.
df = df.drop("全国", axis=0)
df["平成30年"] = pd.to_numeric(df["平成30年"].str.replace(",", ""))
df = df.sort_values("平成30年",ascending=False)
df["平成30年"].plot.bar(figsize=(10, 6))
plt.show()
```

```python
# Show population increase/decrease
# Create and display a bar chart with column data for the year 2008.
df["平成29年"] = pd.to_numeric(df["平成29年"].str.replace(",", ""))
df["人口増減"] = df["平成30年"] - df["平成29年"]
df = df.sort_values("人口増減",ascending=False)
df["人口増減"].plot.bar(figsize=(10, 6))
plt.show()
```

In the graph above, the maximum value of memory on the y-axis is 100, although it appears to be concentrated in Tokyo. Since the maximum value of memory on the Y-axis so far has been 14,000, the scale should be adjusted.

```python
df["人口増減"].plot.bar(figsize=(10, 6))
plt.ylim(-40, 14000)
plt.show()
```

On this scale, little variation is content.
